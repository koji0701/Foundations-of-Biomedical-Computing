```python
# First import Biopython
import Bio
```

Biopython has a class called 'seq', which is sort of like strings in regular python, except it allows common complex operations on biological sequences. We can define a sequence in Bio.Seq as DNA, RNA, or protein. From there we can use dot methods like .complement(), .transcribe(), or .translate().  (Of course we could code the mapping of codons into proteins from scratch, but Bio does that work for us.) Run the following example code:

```python
# example
from Bio.Seq import Seq

my_gene = Seq("ATGGCCTTAAATCGGTTACA")
print(my_gene)
print(my_gene.complement())
print(my_gene.transcribe())
print(my_gene.transcribe().translate())

# what happens if the Seq has an incorrect element?
my_gene = Seq("ATGGCCTTAQATCGGTTACA")
print(my_gene)
print(my_gene.complement())
print(my_gene.transcribe())
print(my_gene.transcribe().translate())
```

## Working with sequences
Write a formal function using *def* to build and work with gene sequences. 

First, either use the desired length as the imput argument to the funtions, or from within the function, query the user to enter an integer number greater than 10 representing the desired length of sequence.

Hint: a user entry can be placed in a variable using the following syntax:

Next, use *random* methods as discussed in lecture to generate a DNA sequence of the length entered. Be sure that your sequence is made from only letters A, T, C, and G. Your DNA sequence must end up as a string, as shown in the sample code above. You can look up how to turn a list of strings into a single string.

The function should *return* the DNA, its complement, its RNA transcription, and the translated proteins, with text that indicates each one.

The function must recover gracefully from the following errors:
* the user enters something that is not an integer or the input argument is not an integer
* the integer is not greater than ten
* the integer is not a multiple of three




```python
import random
from Bio.Seq import Seq

def geneSequences():
    while True:
        try:
            lengthOfSequence = int(input("Enter the desired length of a sequence (>10 and a multiple of 3): "))
            if (lengthOfSequence < 11): 
                print("Invalid input, please put a number greater than 10")
            elif (lengthOfSequence % 3 != 0):
                print("Invalid input, please put a number that is a multiple of 3")
            else:
                break
        except:
            print("Please type an integer")
    letters = "ATCG"
    DNAstr = ""
    rand = random.randint(0,3)
    while len(DNAstr) < lengthOfSequence:
        DNAstr += letters[rand]
        rand = random.randint(0,3)
    DNA = Seq(DNAstr)
    complement = DNA.complement()
    RNA = DNA.transcribe()
    protein = RNA.translate()

    return {
        "DNA": DNA, 
        "complement": complement,
        "RNA": RNA,
        "protein": protein
    }

rv = geneSequences()
print(rv)


    
```
